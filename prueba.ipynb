```python
import spacy
nlp = spacy.load('es_core_news_sm')
```

```python
text = "El ingeniero Tino Mosu estudia en la Universidad Tecnológica Nacional (UTN). Actualmente reside en Buenos Aires y busca liderar proyectos de migración Cloud y automatización DevOps. En su día a día, domina lenguajes como Python y Java, además de administrar entornos en Linux utilizando Git y GitHub. Para propuestas laborales, los reclutadores pueden escribir a t.mosu_dev@email.com o comunicarse al teléfono +54 9 11 9876-5432 a partir del 15 de abril."
text
```

```python
doc = nlp(text)
doc
```

```python
for token in doc:
    print(token)
```

```python
for token in doc:
    if token.pos_ == "VERB":
        print(token)
        
```

```python
for entidad in doc.ents:
    print(entidad.text, entidad.label_)
```
