```python
import spacy
```

```python
nlp = spacy.load("es_core_news_sm") #cargo el idioma
```

```python
text = "El 5 de abril de 2024, Juan Pérez envió un correo a juan.perez@gmail.com desde Buenos Aires.En el mensaje dijo que trabaja como programador y que su número de teléfono es +54 9 11 1234-5678.También mencionó que su página web es https://juanprograma.com"
text
```

```python
doc = nlp(text) #cargo el idioma y le paso el texto que tengo como parametro. Tokenizacion interna
doc
```

```python
for token in doc:
    print(token) #extraigo cada token uno por uno
```

```python
for token in doc:
    if token.pos_ == "VERB" : #Busco los verbos en esa lista de token
        print(token)
```

```python
for entidad in doc.ents:  #Entidades
    print(entidad.text, entidad.label_)
```
